```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
np.random.seed(42)
X = np.random.randn(500, 1)
y = 1 + 2*X + 1.2*np.random.randn(500, 1)
```

```python
print(X.shape)
print(y.shape)
```

```python
%matplotlib inline
plt.figure(figsize=(10, 8))
plt.scatter(X, y, color='red', marker='o', alpha=0.5);
```

```python
def theta(X, y):
    m = X.shape[0]
    X = np.concatenate([X, np.ones((m, 1))], axis=1)
    y = y.reshape((m, 1))
    return np.dot(np.linalg.inv(np.dot(X.T, X)), np.dot(X.T, y))
```

```python
def predict(X, theta):
    m = X.shape[0]
    X = np.concatenate([X, np.ones((m, 1))], axis=1)
    prediction = np.dot(X, theta)
    return prediction
```

```python
normal_theta = theta(X, y)
```

```python
normal_theta
```

```python
preds = predict(X, normal_theta)
```

```python
%matplotlib inline
plt.figure(figsize=(10, 8))
plt.scatter(X, y, color='red', marker='o', alpha=0.5)
plt.plot(X, preds, color='green', marker='o', alpha=0.6);
```
