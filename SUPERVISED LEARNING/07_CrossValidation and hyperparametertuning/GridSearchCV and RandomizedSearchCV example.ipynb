# GridSearchCV and RandomizedSearchCV on digits dataset

**IMPORTING LIBRIRIES**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn import svm
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

**IMPORTING DATASET**

```python
from sklearn.datasets import load_digits
digits = load_digits()
```

```python
dir(digits)
```

```python
from sklearn.model_selection import train_test_split

train_x, test_x, train_y, test_y = train_test_split(digits.data, digits.target)
```

**APPLYING DIFFERENT MODELS**

```python
log_model = LogisticRegression()
svm_model = SVC(kernel='rbf', random_state=42)
tree_model = DecisionTreeClassifier(criterion='gini')

log_model.fit(train_x, train_y)
svm_model.fit(train_x, train_y)
tree_model.fit(train_x, train_y)

log_pred = log_model.predict(test_x)
svm_pred = svm_model.predict(test_x)
tree_pred = tree_model.predict(test_x)

print(log_model.score(train_x, train_y))
print(svm_model.score(train_x, train_y))
print(tree_model.score(train_x, train_y))

print(log_model.score(test_x, test_y))
print(svm_model.score(test_x, test_y))
print(tree_model.score(test_x, test_y))
```

**MODELS HYPERPARAMETERS SELECTION**

```python
model_params = {
    'logistic_regression': {
        'model': LogisticRegression(),
        'params': {
            'C': [0.5, 1, 1.5, 2, 2.5, 3, 5, 10]
        }
    }, 
    'svm': {
        'model': SVC(gamma='auto', random_state=42),
        'params': {
            'kernel': ['linear', 'rbf', 'poly'],
            'C': [1, 5, 10, 20, 25, 50, 100]
        }
    }, 
    'tree': {
        'model': DecisionTreeClassifier(random_state=42),
        'params': {
            'criterion': ['gini', 'entropy'],
            'max_depth': [2, 4, 6, 8]
        }
    },
    'random_forest': {
        'model': RandomForestClassifier(n_jobs=-1, random_state=42),
        'params': {
            'n_estimators': [10, 20, 30, 40, 50, 75, 80, 100, 120, 130, 150],
            'criterion': ['gini', 'entropy'],
            'bootstrap': [True, False],
            'max_depth': [2, 4, 6, 8, 10]
        }
    }
}
```

## Training using GridSearchCV and RandomizedSearchCV

```python
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
```

**GridSearchCV**

```python
%%time
scores = []

for model in model_params.values():
    clf = GridSearchCV(model['model'], model['params'], cv=10, return_train_score=False)
    clf.fit(digits.data, digits.target)
    scores.append({'Model': model['model'], 'Parameters': clf.best_params_, 'Test score': clf.best_score_})
    
df = pd.DataFrame(scores)
```

```python
df
```

**RandomizedSearchCV**

```python
%%time
r_scores = []

for model in model_params.values():
    clf = RandomizedSearchCV(model['model'], model['params'], cv=10, return_train_score=False, n_iter=10)
    clf.fit(digits.data, digits.target)
    r_scores.append({'Model': model['model'], 'Parameters': clf.best_params_, 'Test score': clf.best_score_})
    
df2 = pd.DataFrame(r_scores)
```
